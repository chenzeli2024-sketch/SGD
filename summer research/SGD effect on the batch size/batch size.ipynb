```python
"""
Import the data with Isotropic model
"""



import sys
sys.path.append("..")
import numpy as np
import tqdm
import pickle
from tqdm import tqdm
from tqdm import trange
import pickle
iso=__import__("Isotropic features")
isom=iso.isomodel()
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

A,x_0,x_1,beta,r,n,d=isom.generate()         # import the data
b=np.dot(A,x_1)
```

```python
beta=range(1,41)                                     # the range of batch size beta

def SGD(x,beta,s):                                     # define SGD iteration
    pos=np.random.choice(range(1000),beta,False)
    Pk=np.zeros((n,n))
    Pk[pos,pos]=1
    x_new=x-s/n*np.dot(A.T.dot(Pk),(A.dot(x)-b))
    
    return x_new
```

```python
"""
The process iterates SGD model with 1000 times for each batch size
"""
ss=(2/(r*np.sum(np.diag(np.dot(A.T,A)/n))/d))/1.2       # step size
x1=x_0
epoch=1000
err_1=[]
for j in tqdm(beta):
    err=[]
    for i in range(epoch):
        x1=SGD(x1,j,ss)
        err.append(np.sum((A.dot(x1)-b)**2)/(2*n))
    err_1.append(err)
```

```python
"""
To store and load the data with pickle
"""

filek=open(".\\data-1p2.pickle","wb")
pickle.dump(err_1,filek)
filek.close()
```

```python
"""We only record the value of f(x) at 400th iteration for each step size under different step size
"""


set1=[400]                           # set the number of iterations as 400                           

turn1=[]                             # record the value of f(x) for each batch size at 400th iteration with the specific step size
for k in set1:
    temp=[]
    for i in err_1:
        temp.append(i[k-1])
    turn1.append(temp)
    
turn2=[]
for k in set1:
    temp=[]
    for i in err_2:
        temp.append(i[k-1])
    turn2.append(temp)
    
turn3=[]
for k in set1:
    temp=[]
    for i in err_3:
        temp.append(i[k-1])
    turn3.append(temp)
    
turn4=[]
for k in set1:
    temp=[]
    for i in err_4:
        temp.append(i[k-1])
    turn4.append(temp)
```
