```python
"""
Import data of isotropic model and necessary class 
to conduct ridge regression under SGD model
"""

import sys
sys.path.append("..")
import numpy as np
import pickle
rid=__import__("Ridge regression")                  # ridge regression
iso=__import__("Isotropic features")                # data for isotropic model
isom=iso.isomodel()
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

A,x_0,x_1,beta,r,n,d=isom.generate()                # generate data
```

```python
ss=(2/(r*np.sum(np.diag(np.dot(A.T,A)/n))/d))/2                   # calculate the step size
ridm=rid.ridge(A,x_0,x_1,beta,r,n,d,ss) 
la=list(np.around(np.linspace(0,1,100,endpoint=False)
                               ,decimals=6))

errr=[]
b=A.dot(x_1)
for lambd in la:
    x_=x_0
    x=x_0
    err=[]
    for i in range(100):
        err.append(np.sum((A.dot(x_)-b)**2)/
                                   (2*n)+lambd*np.sum(x_**2)/2)
        x=ridm.SGDs(x_,lambd)
        x_=x
        
    errr.append(err)
```

```python
mini=[]
for i in errr:
    mini.append(i[-1])
plt.plot(la,mini)
```

```python
"""
Calculate the ridge regression under SGD model with update stops 
when k satisfies the criteria mentioned in the report when step size 
is \gamma=\gamma_{max}/2
"""
ss=(2/(r*np.sum(np.diag(np.dot(A.T,A)/n))/d))/2                   # calculate the step size
ridm=rid.ridge(A,x_0,x_1,beta,r,n,d,ss)                           # input the data to the ridge regression
stable,collapse=ridm.iterate_dls()                                # stable stores the end values of f(x) 
                                                                  # when iteration stops for each lambda
                                                                  # collapse stores the lambda s which cause the divergence of SGD model

filek=open(".\\data\\nosaepoch\\1.pickle","wb")                   # store and load the results with pickle
pickle.dump(stable,filek)
filek.close()

file1=open(".\\data\\nosaepoch\\2-b.pickle","wb")
pickle.dump(collapse,file1)
file1.close()

"""
Calculate the ridge regression under SGD model with update stops 
when k =100 when step size is \gamma=\gamma_{max}/2
"""
stable_,collapse_=ridm.iterate_se()                               # stable_ stores the end values of f(x) 
                                                                  # when iteration stops for each lambda
                                                                  # collapse_ stores the lambda s which cause the divergence of SGD model

filek=open(".\\data\\sameepoch\\2.pickle","wb")
pickle.dump(stable_,filek)
filek.close()

file1=open(".\\data\\sameepoch\\2-b.pickle","wb")
pickle.dump(collapse_,file1)
file1.close()
```
