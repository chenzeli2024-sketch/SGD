```python
import pickle
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
"""
Import results of numerical solution of volterra equation with pickle
"""

file1=open("data\\Volterra\\2.pickle","rb")           
file2=open("data\\Volterra\\1p2.pickle","rb")
file3=open("data\\Volterra\\4.pickle","rb")
file4=open("data\\Volterra\\8.pickle","rb")

y1=pickle.load(file1)
y2=pickle.load(file2)
y3=pickle.load(file3)
y4=pickle.load(file4)
```

```python
"""
Import results of SGD model with pickle
"""

file1=open("data\\SGD\\sgd-2.pickle","rb")
file2=open("data\\SGD\\sgd-1p2.pickle","rb")
file3=open("data\\SGD\\sgd-4.pickle","rb")
file4=open("data\\SGD\\sgd-8.pickle","rb")

err1=pickle.load(file1)
err2=pickle.load(file2)
err3=pickle.load(file3)
err4=pickle.load(file4)
```

```python
"""
Import results of analytic solution of volterra equation with pickle
"""

file1=open("data\\analytic\\2.pickle","rb")
file2=open("data\\analytic\\1p2.pickle","rb")
file3=open("data\\analytic\\4.pickle","rb")
file4=open("data\\analytic\\8.pickle","rb")

y_1=pickle.load(file1)
y_2=pickle.load(file2)
y_3=pickle.load(file3)
y_4=pickle.load(file4)
```

```python
"""Plot the results
"""

fig=plt.figure(figsize=(9,7.5))


ax1=plt.subplot(2,2,1)
ax1.set_yscale("log")
ax1.set_title(r"$\gamma=\gamma_{max}/1.2$")
ax1.plot(range(1,41),y2,color="lime",label="Numerical solution of Volterra",marker="p")
ax1.plot(range(1,41),err2,color="deepskyblue",label="SGD",marker="2")
ax1.plot(range(1,41),y_2,color="darkviolet",label="Analytic solution of Volterra",marker="+")


ax2=plt.subplot(2,2,2)
ax2.set_yscale("log")
ax2.set_title(r"$\gamma=\gamma_{max}/2$")
ax2.plot(range(1,41),y1,color="lime",label="Numerical solution of Volterra",marker="p")
ax2.plot(range(1,41),err1,color="deepskyblue",label="SGD",marker="2")
ax2.plot(range(1,41),y_1,color="darkviolet",label="Analytic solution of Volterra",marker="+")

ax3=plt.subplot(2,2,3)
ax3.set_yscale("log")
ax3.set_title(r"$\gamma=\gamma_{max}/4$")
ax3.plot(range(1,41),y3,color="lime",label="Numerical solution of Volterra",marker="p")
ax3.plot(range(1,41),err3,color="deepskyblue",label="SGD",marker="2")
ax3.plot(range(1,41),y_3,color="darkviolet",label="Analytic solution of Volterra",marker="+")

ax4=plt.subplot(2,2,4)
ax4.set_yscale("log")
ax4.set_title(r"$\gamma=\gamma_{max}/8$")
ax4.plot(range(1,41),y4,color="lime",label="Numerical solution of Volterra",marker="p")
ax4.plot(range(1,41),err4,color="deepskyblue",label="SGD",marker="2")
ax4.plot(range(1,41),y_4,color="darkviolet",label="Analytic solution of Volterra",marker="+")

fig.text(0.5,0,"epoch",ha="center")
fig.text(0,0.5,r"$f(x)$",va="center",rotation="vertical")

lines, labels = ax4.get_legend_handles_labels()
    
fig.legend(lines, labels, loc = 'upper center')
plt.tight_layout()
fig.savefig("plot\\SGD+v.pdf",dpi=600,format="pdf")
```
